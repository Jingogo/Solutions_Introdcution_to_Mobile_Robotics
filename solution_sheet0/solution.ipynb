```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import stats 
```

### Exercise 1: Defining functions 

```python
def firstFunction(x):
  return math.cos(x)*math.exp(x) 
```

```python
### Exercise 2: Plotting data
```

```python
x = np.linspace(-2*np.pi, 2*np.pi)
y = [firstFunction(i) for i in x]
plt.plot(x,y)
plt.title("f(x)=cos(x)exp(x)")
plt.savefig("plot_fuction.png")
```

### Exercise 3: Generating random numbers

```python
np.random.seed(31)
mu = 0.5
sigma = 2.0
gaussian_samples = np.random.normal(mu,sigma,100000)
mean = np.mean(gaussian_samples)
std_dev = np.std(gaussian_samples)

print("Gaussian distribution")
print("mean: expected == actual mean?", abs(mean-mu)<0.01)
print("standard deviation: expected == actual mean?", abs(std_dev-sigma)<0.01) 

a = 0
b = 10
uniform_samples = np.random.uniform(a, b, 1000000)
mean = np.mean(uniform_samples)
std_dev = np.std(uniform_samples)
expected_mean = 5
expected_std_dev = 5*np.sqrt(3)/3
print("\nUniform distribution")
print("mean: expected == actual mean?", abs(mean-expected_mean)<0.01)
print("standard deviation: expected == actual mean?", abs(std_dev-expected_std_dev)<0.01) 
```
